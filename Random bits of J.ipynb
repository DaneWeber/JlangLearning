# Ensure that every box contains a one-dimensional string.

```j
(([:1:' '&,) ::0: * (1=#@$))@> 'aaa' ; 'bb'
```

```j
(([:1:' '&,) ::0: * (1=#@$))@> 'aaa' ; 'bb' ; (2 2 $ 'cccc') ; i. 4
```

```j
*./ (([:1:' '&,) ::0: * (1=#@$))@> 'aaa' ; 'bb' ; (2 2 $ 'cccc') ; i. 4
```

```j
*./ (([:1:' '&,) ::0: * (1=#@$))@> 'aaa' ; 'bb'
```

```j
(([:1:' '&,) ::0:)@> 'aaa' ; 'bb' ; (2 2 $ 'cccc') ; i. 4
```

' '&, will fail on non-strings, and thus 1: ' '&, will also fail. :: 0: returns a 0 if the other verbs fails.

```j
((1:' '&,) ::0:)@> 'aaa' ; 'bb' ; (2 2 $ 'cccc') ; i. 4
```
